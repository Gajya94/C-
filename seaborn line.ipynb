```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame(dict(time=np.arange(500),value=np.random.randn(500)))
new=sns.relplot(x="time",y="value",data=df)
new.fig.autofmt_xdate()
```

```python
new=sns.relplot(x="time",y="value",kind="line",data=df)
new.fig.autofmt_xdate()
```

```python
df = pd.DataFrame(dict(time=np.arange(500),value=np.random.randn(500).cumsum()))
new=sns.relplot(x="time",y="value",kind="line",data=df)
new.fig.autofmt_xdate()
```

```python
fm=sns.load_dataset("titanic")
print(fm)
```

```python
sns.relplot(x="survived",y="sex",data=fm)
```
