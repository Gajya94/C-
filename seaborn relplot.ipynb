```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
sns.set(style="darkgrid")
```

```python
tips = sns.load_dataset("tips")
print(tips)
```

```python
sns.relplot(x="total_bill",y="tip",data=tips)
```

```python
sns.relplot(x="total_bill",y="tip",hue="smoker",style="smoker",data=tips)
```

```python
sns.relplot(x='total_bill',y='tip',hue="size",data=tips)
```

```python
sns.relplot(x="total_bill",y="tip",hue="size",data=tips,palette="ch:r=-.5, l= .75")
# hue is for color
#smoker is for
```

```python
sns.relplot(x="total_bill",y="tip",hue="time",style="smoker",data=tips)
```

```python
sns.relplot(x="total_bill",y="tip",hue="size",size="size",data=tips)
#hue is for 1 color in grediant form
#size is for changing the size of the plot according to the datapath
```

```python
sns.relplot(x="total_bill",y="tip",hue="size",size="size",sizes=(100,280),data=tips)
```
